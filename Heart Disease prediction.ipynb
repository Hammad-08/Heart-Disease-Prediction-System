```python
cat_val = []
cont_val = []
for column in Heart_data.columns:
    if Heart_data[column].nunique()<=10:
        cat_val.append(column)
    else:
        cont_val.append(column)
           
        
```

```python
cat_val.remove("sex")
cat_val.remove("target")
Heart_data=pd.get_dummies(Heart_data,columns=cat_val,drop_first=True)
```

```python
st = StandardScaler()
Heart_data[cont_val]=st.fit_transform(Heart_data[cont_val])
Heart_data.head()
```

# GUI

```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd

# Function to remove default value when clicked
def on_click(event):
    entry = event.widget
    if entry.get() == "0":
        entry.delete(0, tk.END)

# Function to make prediction and show result
def predict_disease():
    new_data = {
        'age': [int(entry_widgets["age"].get())],
        'sex': [int(entry_widgets["sex"].get())],
        'cp': [int(entry_widgets["cp"].get())],
        'trestbps': [int(entry_widgets["trestbps"].get())],
        'chol': [int(entry_widgets["chol"].get())],
        'fbs': [int(entry_widgets["fbs"].get())],
        'restecg': [int(entry_widgets["restecg"].get())],
        'thalach': [int(entry_widgets["thalach"].get())],
        'exang': [int(entry_widgets["exang"].get())],
        'oldpeak': [float(entry_widgets["oldpeak"].get())],
        'slope': [int(entry_widgets["slope"].get())],
        'ca': [int(entry_widgets["ca"].get())],
        'thal': [int(entry_widgets["thal"].get())]
    }
    new_data_df = pd.DataFrame(new_data)
    prediction = rf.predict(new_data_df)
    if prediction[0] == 0:
        messagebox.showinfo("Prediction", "No disease")
    else:
        messagebox.showinfo("Prediction", "Disease")

# Create GUI window
root = tk.Tk()
root.title("Heart Disease Prediction")
root.geometry("400x500")

# Style
root.configure(bg='#B3E5FC')  # Light blue background
style = ttk.Style()
style.configure('TButton', foreground='black', background='#FF5722', font=('Arial', 12, 'bold'))  # Orange button with black text
style.configure('TLabel', foreground='#000000', background='#B3E5FC', font=('Arial', 12, 'bold'))

# Add input fields
input_fields = [
    "age", "sex", "cp", "trestbps", "chol", 
    "fbs", "restecg", "thalach", "exang", "oldpeak", 
    "slope", "ca", "thal"
]

entry_widgets = {}

for i, field in enumerate(input_fields):
    ttk.Label(root, text=field.capitalize() + ":", background='#B3E5FC').place(x=20, y=20+i*30)
    entry = ttk.Entry(root, font=('Arial', 12))
    entry.insert(0, "0")  # Insert default value
    entry.bind("<FocusIn>", on_click)  # Remove default value when clicked
    entry.place(x=150, y=20+i*30)
    entry_widgets[field] = entry

# Prediction button
predict_button = ttk.Button(root, text="Predict", command=predict_disease)
predict_button.place(x=150, y=20+len(input_fields)*30+20)

# Run GUI
root.mainloop()
```
